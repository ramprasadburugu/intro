Let's build a simple program to calculate age of all student's in a class

```python
from datetime import datetime

students = [
    {"name": "Noah", "dob": "05/15/2008"},
    {"name": "Olivia", "dob": "11/30/2007"},
    {"name": "Liam", "dob": "01/20/2009"},
]

current_date = datetime.now()

for student in students:
    date_of_birth = datetime.strptime(student["dob"], "%m/%d/%Y")
    age_in_years = current_date.year - date_of_birth.year
    if (current_date.month, current_date.day) < (date_of_birth.month, date_of_birth.day):
        age_in_years -= 1
    print(f"{student['name']} is {age_in_years} years old.")
```

```python
from datetime import datetime

def calculate_age(date_of_birth, current_date):
    """Calculates age in years based on date of birth and current date."""
    age_in_years = current_date.year - date_of_birth.year
    if (current_date.month, current_date.day) < (date_of_birth.month, date_of_birth.day):
        age_in_years -= 1
    return age_in_years

students = [
    {"name": "Noah", "dob": "05/15/2008"},
    {"name": "Olivia", "dob": "11/30/2007"},
    {"name": "Liam", "dob": "01/20/2009"},
]

current_date = datetime.now()

for student in students:
    date_of_birth = datetime.strptime(student["dob"], "%m/%d/%Y")
    age = calculate_age(date_of_birth, current_date)
    print(f"{student['name']} is {age} years old.")
```

```python
from datetime import datetime

def calculate_age(date_of_birth, current_date):
    """Calculates age in years based on date of birth and a specified date."""
    age_in_years = current_date.year - date_of_birth.year
    if (current_date.month, current_date.day) < (date_of_birth.month, date_of_birth.day):
        age_in_years -= 1
    return age_in_years

students = [
    {"name": "Noah", "dob": "05/15/2008"},
    {"name": "Olivia", "dob": "11/30/2008"},
    {"name": "Liam", "dob": "01/20/2009"},
]

# Calculate age for a current date
current_date = datetime.now()

for student in students:
    date_of_birth = datetime.strptime(student["dob"], "%m/%d/%Y")
    age = calculate_age(date_of_birth, current_date)
    print(f"{student['name']} is {age} years old.")

# Calculate age for a future date
future_date_str = "06/01/2030"  # Example future date
future_date = datetime.strptime(future_date_str, "%m/%d/%Y")

print("\nCalculating age in the future (", future_date_str, "):")

for student in students:
    date_of_birth = datetime.strptime(student["dob"], "%m/%d/%Y")
    age_in_future = calculate_age(date_of_birth, future_date)
    print(f"{student['name']} will be {age_in_future} years old.")
```

## Group data

### Subtask:
Group the `df_cars` DataFrame by both 'Make' and 'Fuel' and count the number of vehicles for each combination.


**Reasoning**:
Group the DataFrame by 'Make' and 'Fuel', sum the 'Vehicles', and reset the index to create the `df_grouped_cars` DataFrame.



```python
df_grouped_cars = df_cars.groupby(['Make', 'Fuel'])['Vehicles'].sum().reset_index()
display(df_grouped_cars.head())
```

## Create an interactive chart

### Subtask:
Use an interactive plotting library (like Altair) to generate a chart that visualizes the relationship between 'Make' and 'Fuel' based on the vehicle counts. A stacked bar chart or a heatmap could be suitable options.


**Reasoning**:
Create a stacked bar chart using Altair to visualize the distribution of vehicles by make and fuel type.



```python
import altair as alt

chart = alt.Chart(df_grouped_cars).mark_bar().encode(
    x=alt.X('Make', axis=None),
    y=alt.Y('Vehicles'),
    color='Fuel',
    tooltip=['Make', 'Fuel', 'Vehicles']
).properties(
    title='Vehicle Count by Make and Fuel Type'
).interactive()

chart
```

## Obtain california geographical data

### Subtask:
Find and load geographical data for California, such as a GeoJSON or shapefile, that includes boundaries for ZIP codes or other suitable administrative divisions.


**Reasoning**:
Search for California ZIP code GeoJSON data and load it into a GeoDataFrame.



```python
import geopandas as gpd

# Attempt to find and load California ZIP code GeoJSON data.
# Searching for publicly available data sources.
# A common source for geographic data is the U.S. Census Bureau, but their data might require significant processing.
# Open source geographic data repositories like OpenStreetMap or specific state/local government portals are other options.
# Let's assume we found a publicly available GeoJSON URL for California ZIP codes.
# Note: The following URL is a placeholder and might not be valid. A real-world scenario would require finding a valid data source.
geojson_url = "https://raw.githubusercontent.com/OpenDataDE/State-zip-code-GeoJSON/master/ca_california_zip_codes_geo.min.json" # Placeholder URL

try:
    gdf_california_zip = gpd.read_file(geojson_url)
    print("Successfully loaded GeoJSON data.")
    display(gdf_california_zip.head())
except Exception as e:
    print(f"Error loading GeoJSON data: {e}")
    print("Could not load the geographical data. Please ensure the URL is correct and the data is accessible.")
    gdf_california_zip = None
```

## Process car data

### Subtask:
Ensure the `df_cars` DataFrame is ready for merging, potentially aggregating car counts by ZIP code if not already done.


**Reasoning**:
Examine the structure and content of the df_cars DataFrame to understand its current state.



```python
print(df_cars.info())
display(df_cars.head())
```

**Reasoning**:
Group the `df_cars` DataFrame by 'ZIP Code', sum the 'Vehicles' column, rename the aggregated column, and display the head to verify the changes.



```python
df_cars_by_zip = df_cars.groupby('ZIP Code')['Vehicles'].sum().reset_index()
df_cars_by_zip = df_cars_by_zip.rename(columns={'Vehicles': 'Total Vehicles'})
display(df_cars_by_zip.head())
```

## Merge geographical and car data

### Subtask:
Combine the geographical data with the car count data based on a common identifier (like ZIP code).


```python
gdf_merged = gdf_california_zip.merge(df_cars_by_zip, how='left', left_on='ZCTA5CE10', right_on='ZIP Code')
print(gdf_merged.info())
display(gdf_merged.head())
```

## Create the heatmap

### Subtask:
Use a suitable library (like GeoPandas and Matplotlib, or Altair with a GeoJSON) to generate a heatmap of California, where the color intensity represents the car count in each area.


**Reasoning**:
Plot the merged GeoDataFrame `gdf_merged` as a heatmap using matplotlib.
